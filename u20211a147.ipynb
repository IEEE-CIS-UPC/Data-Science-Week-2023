###**Datos Atípicos (Outliers)** (1.5 pts)

####**Limpieza de los datos atípicos de la columna Fare** (1.5 pts)
**En esta sección deberas implementar el código correspondiente para obtener el output esperado para cada una de las siguientes preguntas:**
1. Muestra las observaciones que contengan los outliers en la columna Fare
2. Implementa una función para reducir los datos atípicos de esta columna

```python
#Responde aquí la pregunta número 1:
#hallamos la mediana y el rango intercuartil
mediana_Fare = df["Fare"].median()
rango_IC =df["Fare"].quantile(0.75) - df["Fare"].quantile(0.25)
#definir cual sera el limite para saber si es outliers
limite = mediana_Fare + rango_IC
#filtar
df_outliers = df[df["Fare"]>limite]
#mostrar
df_outliers
```

```python
#Responde aquí la pregunta número 2:
def solveFare(df):
    mediana_Fare = df["Fare"].median()
    rango_IC =df["Fare"].quantile(0.75) - df["Fare"].quantile(0.25)
    limite = mediana_Fare + rango_IC
    df["Fare"] = df["Fare"].clip(lower=limite)
    return df
    
```
